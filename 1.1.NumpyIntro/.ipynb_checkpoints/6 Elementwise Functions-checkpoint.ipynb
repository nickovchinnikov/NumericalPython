# Elementwise Functions

```python
import numpy as np
```

```python
x = np.linspace(-1, 1, 11)
x
```

### np.sin example

```python
y = np.sin(np.pi * x)
y
```

```python
np.round(y, decimals=4)
```

### probably better example of np.sin

```python
# multipliers of pi

x = np.pi * np.array([0, 1/2, 1, 3/2, 2])
x
```

```python
# the same with linspace

x = np.pi * np.linspace(0, 2, 5)
x
```

```python
y = np.sin(x)
y
```

```python
# Round result data

np.round(y, decimals=4)
```
