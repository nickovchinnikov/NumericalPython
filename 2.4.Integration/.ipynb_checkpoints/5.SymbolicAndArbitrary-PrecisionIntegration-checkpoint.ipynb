# Symbolic and Arbitrary-Precision Integration

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import integrate
import sympy
import mpmath
sympy.init_printing()
```

```python
x = sympy.symbols("x")
f = 2 * sympy.sqrt(1-x**2)
a, b = -1, 1

val_sym = sympy.integrate(f, (x, a, b))

# The result of integration
val_sym
```

```python
mpmath.mp.dps = 75
```

```python
f_mpmath = sympy.lambdify(x, f, 'mpmath')

val = mpmath.quad(f_mpmath, (a, b))

# The result is pi
sympy.sympify(val)
```

```python
sympy.N(val_sym, mpmath.mp.dps+1) - val
```

```python
x, y, z = sympy.symbols("x, y, z")

f2 = sympy.cos(x) * sympy.cos(y) * sympy.exp(-x**2 - y**2)
f3 = sympy.cos(x) * sympy.cos(y) * sympy.cos(z) * sympy.exp(-x**2 - y**2 - z**2)

f2_mpmath = sympy.lambdify((x, y), f2, 'mpmath')
f3_mpmath = sympy.lambdify((x, y, z), f3, 'mpmath')
```

The integrals can then be evaluated to the desired accuracy by setting `mpmath.mp.dps` and calling `mpmath.quad`

```python
mpmath.mp.dps = 30
mpmath.quad(f2_mpmath, (0, 1), (0, 1))
```

#### Very slow!

```python
res = mpmath.quad(f3_mpmath, (0, 1), (0, 1), (0, 1))
sympy.sympify(res)
```

Again, this gives access to levels of accuracy that is beyond what `scipy.integrate.quad` can achieve, but this additional accuracy comes with a hefty increase in
computational cost. Note that the type of the object returned by `mpmath.quad` is a
multiprecision `float(mpf)`. It can be cast into a SymPy type using `sympy.sympify`
