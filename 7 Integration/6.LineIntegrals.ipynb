# Line Integrals

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy import integrate
import sympy
import mpmath
sympy.init_printing()
```

```python
t, x, y = sympy.symbols("t, x, y")

# Create a parametric curve between [0, 2pi]
C = sympy.Curve([sympy.cos(t), sympy.sin(t)], (t, 0, 2 * sympy.pi))
C
```

Once the integration path is specified, we can easily compute the corresponding
line integral for a given integrand using line_integral. For example, with the integrand
$f (x, y) = 1$, the result is the circumference of the unit circle

```python
# Area of the unit circle
sympy.line_integrate(1, C, [x, y])
```

The result is less obvious for a nontrivial integrand, such as in the following example
where we compute the line integral with the integrand $f (x, y) = x^2y^2$:

```python
sympy.line_integrate(x**2 * y**2, C, [x, y])
```
