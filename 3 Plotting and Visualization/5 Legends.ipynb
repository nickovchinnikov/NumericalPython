# Legends

## Example based on Getting started first graph

```python
%matplotlib inline
# %config InlineBackend.figure_format='svg'

import matplotlib.pyplot as plt
import matplotlib as mpl

from mpl_toolkits.mplot3d.axes3d import Axes3D

import numpy as np
```

## Example 1

```python
# Data for graph
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x+10

# You can change loc pos
def generate_plot(loc: int):
    # Example of a simple graph created with Matplotlib
    fig, ax = plt.subplots()
    ax.plot(x, y1, color="blue", label="y(x)")
    ax.plot(x, y2, color="red", label="y'(x)")
    ax.plot(x, y3, color="green", label="y‚Äù(x)")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend(loc=loc)
```

### Counter-clockwise direction for the legend square when put 1, 2, 3, 4 to param loc

```python
generate_plot(1),generate_plot(2),generate_plot(3),generate_plot(4)
```

## Example 2

### Legend displayed outside the Axes object and shown with four columns instead of the single one, here using ax.legend(ncol=4, loc=3, bbox_to_anchor=(0, 1))

### Define base colors

```python
base_colors = list(mpl.colors.BASE_COLORS)

# exclude white
base_colors_len = len(base_colors) - 1

base_colors
```

### Data for graph 2

```python
# range [-1; 1]
x = np.array(np.linspace(-1, 1, 8))
x
```

```python
def generate_rays(x):
    fig, ax = plt.subplots()
    # Generate subplots
    for i in range(len(x)):
        y = i*x
        color = base_colors[i % base_colors_len]
        ax.plot(x, y, color, label=f"$y_{i}(x) = {i}x$")
        ax.legend(ncol=4, loc=3, bbox_to_anchor=(0, 1))
        
generate_rays(x)
```

```python
# range [-10; 10]
x = np.array(np.linspace(-10, 10, 31))
x
```

```python
# very tight picture

generate_rays(x)
```
