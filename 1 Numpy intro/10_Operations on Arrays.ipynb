# Operations on Arrays

```python
import numpy as np
```

```python
data = np.arange(9).reshape(3, 3)
data
```

```python
np.transpose(data)
```

```python
data = np.random.randn(1, 2, 3, 4, 5)
data.shape
```

```python
data.T.shape
```

```python
data = np.arange(9).reshape(3, 3)
data
```

### Flip rows with np.fliplr

```python
np.fliplr(data)
```

### Flip cols with np.flipup

```python
np.flipud(data)
```

```python
np.rot90(data)
```

```python
np.sort(data)
```
