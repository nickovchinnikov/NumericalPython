## Arithmetic Operations

### The standard arithmetic operations with NumPy arrays perform elementwise operations

```python
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])
```

```python
x
```

```python
y
```

```python
x + y
```

```python
y - x
```

### Not a dot product!

```python
x * y
```

```python
y / x
```

```python
x * 2
```

```python
2 ** x
```

```python
y / 2
```

```python
(y / 2).dtype
```

```python
x = np.array([1, 2, 3, 4]).reshape(2, 2)
x
```

```python
z = np.array([1, 2, 3, 4])
z
```

## Error, dimension mismatch

```python
x / z
```

### Now we have (2x2) / (1*2)

```python
z = np.array([[2, 4]])
```

### Row by row elementwise division

```python
x / z
```

### Concan z

```python
zz = np.concatenate([z, z], axis=0)
```

```python
x / zz
```

```python
z = np.array([[2], [4]])
z.shape
```

### Col by col elementwise division

```python
x / z
```

### Concatenate [col col]

```python
zz = np.concatenate([z, z], axis=1)
zz
```

### And again elementwise

```python
x / zz
```

## The same effect

```python
## Init
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])
```

```python
x = x + y
x
```

```python
## Reinit
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])
```

```python
x += y
x
```
