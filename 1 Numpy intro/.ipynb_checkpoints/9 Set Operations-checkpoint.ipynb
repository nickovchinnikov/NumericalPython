# Set Operations

```python
import numpy as np
```

```python
a = np.unique([1, 2, 3, 3])
b = np.unique([2, 3, 4, 4, 5, 6, 5])

np.in1d(a, b)
```

```python
1 in a
```

```python
1 in b
```

### All items a in b

```python
np.all(np.in1d(a, b))
```

### The standard set operations

```python
np.union1d(a, b)
```

```python
np.intersect1d(a, b)
```

```python
np.setdiff1d(a, b)
```

```python
np.setdiff1d(b, a)
```
