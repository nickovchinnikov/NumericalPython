# Aggregate Functions

```python
import numpy as np
```

## Normal distr example

```python
data = np.random.normal(size=(15,15))
np.mean(data)
```

```python
data.mean()
```

## Simple data example

```python
data = np.arange(1,10).reshape(3,3)
data
```

## Statistical common functions

```python
data.mean()
```

```python
data.std()
```

```python
data.var()
```

### Data sum examples along different axes

```python
# sum all the items
data.sum()
```

```python
# sum all items along the rows
data.sum(axis=0)
```

```python
# sum all items along the cols
data.sum(axis=1)
```
