> The most powerful features of SymPy’s matrix objects, which distinguish it from,
for example, NumPy arrays, are of course that its elements themselves can be symbolic
expressions. For example, an arbitrary 2x2 matrix can be represented with a symbolic
variable for each of its elements

```python
a, b, c, d = sympy.symbols("a, b, c, d")
M = sympy.Matrix([[a, b], [c, d]])
M
```

```python
# Matrix multiplication - dot product

M * M
```

### vector from symbols

```python
x = sympy.Matrix(sympy.symbols("x_1, x_2"))

# matrix times vector product
M * x
```

```python
p, q = sympy.symbols("p, q")
M = sympy.Matrix([[1, p], [q, 1]])
M
```

```python
b = sympy.Matrix(sympy.symbols("b_1, b_2"))
b
```

> and then use, for example, the LUsolve method to solve the linear equation system

```python
x = M.LUsolve(b)
x
```

```python
# or just inverse
x = M.inv() * b
x
```

# Experements

## Very cyclic matrix

```python
def cyclic(m, n):
    return m % 2 + n % 2

def cyclic_decimal(m, n):
    return 10 * m + n
```

### 2-diagonal matrix

```python
A_cyclic = sympy.Matrix(10, 10, cyclic)
A_cyclic
```

```python
### A_10-marching matrix
```

```python
A_10 = sympy.Matrix(10, 10, cyclic_decimal)
A_10
```

### Lots of cyclic diagonals in different order: main-diagonals, cross-diagonals, rows, cols

```python
A_10.det()
```

```python
A_10.norm()
```

```python
A_10.pinv()
```

```python
A_10.eigenvects()
```

### Numpy numerical example

```python
A_10 = np.fromfunction(cyclic_decimal, (10, 10), dtype=int)
```

### Interesting smoothness in pseudo-inverse

```python
np.array(
    np.round(
    np.abs(np.linalg.pinv(np.array(A_10)) * 100000), 
    decimals=0),
dtype=int)
```

### I can call it "smothness" in pseudo-inverse

### Let's take a look to the diagonalization of a Matrix with eigenvalues and eigenvectors

**V** for eigenvalues

**X** fot eigenvectors

```python
V, X = np.linalg.eig(A_10)
V, X
```

Eigenvalues are complex, because of circulant matric
And you can see the **det=0**

**X_INV** For eigenvectors inverse Matrix

```python
X_INV = np.linalg.inv(X)
X_INV
```

### Restore matrix from eigen diagonalization

```python
np.array(
    np.round(np.real(X @ np.diag(V) @ X_INV), decimals=0),
dtype=int)
```

## First 2 eigenvectors can bring most of the information

```python
DIAG_INT_V = np.diag(
    np.array(np.round(np.real(V), decimals=10), dtype=int)
)
DIAG_INT_V
```

### Restore works fine

```python
np.array(
    np.round(np.real(X @ DIAG_INT_V @ X_INV), decimals=0),
dtype=int)
```

### Orthogonality

```python
np.array(
    np.round(np.real(X @ X_INV), decimals=0),
dtype=int)
```
